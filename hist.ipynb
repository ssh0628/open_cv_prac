```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
 
img = cv.imread('data/squirrel_cls.jpg', cv.IMREAD_GRAYSCALE)
plt.subplot(1, 2, 1); plt.imshow(cv.cvtColor(img, cv.COLOR_GRAY2RGB))
plt.subplot(1, 2, 2); plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
img = cv.imread('./data/home.jpg', cv.IMREAD_GRAYSCALE)
plt.hist(img.ravel(),256,[0,256]); plt.show()
```

```python
img = cv.imread('./data/home.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
