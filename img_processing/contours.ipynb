```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((400, 1000, 3), np.uint8)
cv.rectangle(img, (50, 50), (450, 350), (255, 255, 255), -1)
cv.rectangle(img, (550, 50), (950, 350), (255, 255, 255), -1)
plt.imshow(img)
# cv.imwrite("test.jpg", img)
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

ret, thresh = cv.threshold(gray, 127 ,255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
plt.imshow(gray)
```

```python
cv.drawContours(img, contours, -1, (0,255,0), 3)
```

```python

img = cv.imread("./data/smarties.png")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
_, mask = cv.threshold(gray, 200, 255, cv.THRESH_BINARY)
mask_inv = cv.bitwise_not(mask)

cc = np.ones_like(img)
cc[:] = (180, 100, 100)
cc = cv.cvtColor(cc, cv.COLOR_HSV2RGB)

bg = cv.bitwise_and(cc, cc, mask=mask)
fg = cv.bitwise_and(img, img, mask=mask_inv)
res = cv.add(fg, bg)
plt.imshow(res)

res_gray = cv.cvtColor(res, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(res_gray, 200, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
con_img = cv.drawContours(res, contours, -1, (0,0,0), 3)
plt.imshow(con_img)
```
