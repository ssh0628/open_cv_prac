```python
import cv2 as cv
import numpy as np
```

```python
flags = [i for i in dir(cv) if i.startswith('COLOR_')]
flags
```

```python
img = cv.cvtColor(cv.imread("apple.jpeg"), cv.COLOR_BGR2HSV)
ori = cv.cvtColor(cv.imread("apple.jpeg"),cv.COLOR_BGR2RGB)
```

```python
lower_red1, upper_red1 = np.array([170, 0, 0]), np.array([180, 255, 255])
lower_red2, upper_red2 = np.array([0, 0, 0]), np.array([10, 255, 255])
```

```python
mask1 = cv.inRange(img, lower_red1, upper_red1)
res1 = cv.bitwise_and(img, img, mask=mask1)
mask2 = cv.inRange(img, lower_red2, upper_red2)
res2 = cv.bitwise_and(img, img, mask=mask2)
```

```python
import matplotlib.pyplot as plt
plt.imshow(mask2)
```

```python
plt.imshow(res2)
```
