```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
logo = cv.imread("../data/opencv-logo.png")
logo_hsv = cv.cvtColor(logo, cv.COLOR_BGR2HSV)
```

```python
lower_blue, upper_blue = np.array([110, 0, 0]), np.array([130, 255, 255])
lower_red, upper_red = np.array([0, 0, 0]), np.array([10, 255, 255])
lower_green, upper_green = np.array([30, 0, 0]), np.array([60, 255, 255])
```

```python
red_mask = cv.inRange(logo_hsv, lower_red, upper_red)
red_res = cv.bitwise_and(logo_hsv, logo_hsv, mask=red_mask)

green_mask = cv.inRange(logo_hsv, lower_green, upper_green)
green_res = cv.bitwise_and(logo_hsv, logo_hsv, mask=green_mask)

blue_mask = cv.inRange(logo_hsv, lower_blue, upper_blue)
blue_res = cv.bitwise_and(logo_hsv, logo_hsv, mask=blue_mask)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1, 3, 1)
plt.imshow(red_res)
plt.subplot(1, 3, 2)
plt.imshow(green_res)
plt.subplot(1, 3, 3)
plt.imshow(blue_res)
```
